```julia
include("..//src//Atmosphere.jl");
import .Atmosphere as atm;
```

```julia
h = 0: 10 :25000
```

```julia
T = atm.T.(h);
p = atm.p.(h);
rho = atm.rho.(h);
a = atm.a.(h);
```

```julia
using Plots;
gr()
```

```julia
pT = plot(h, T);
pp = plot(h, p);
prho = plot(h, rho);
pa = plot(h, a);

plot(pT, pp, prho, pa, size=(1000, 800))
```
